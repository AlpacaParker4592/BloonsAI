```python
from pandas import DataFrame
import pandas as pd
import os

database_path = os.path.join(os.getcwd(),"bloons_data.csv")
monkey_data = pd.read_csv(database_path)
monkey_data
```

```python
# 결과가 음수인 것은 제외
unallocated_iterations = monkey_data[(monkey_data['generation'] == 1) & (monkey_data['poppedBloons'] < 0)]['iteration'].tolist()  # 목표하던 만큼의 풍선을 터뜨리지 못한 시도 리스트
allocated_iterations = [i+1 for i in range(5) if not i+1 in unallocated_iterations]  # 목표를 완수한 시도 리스트
refined_data = monkey_data.loc[(monkey_data["iteration"].isin(allocated_iterations))]
refined_data
```

```python
refined_data = monkey_data.loc[(monkey_data["iteration"].isin(allocated_iterations)) & (monkey_data['generation'] == 1)]
max_popped_iterations = refined_data[refined_data['totalPoppedBloons'] == max(refined_data['totalPoppedBloons'].tolist())]['iteration'].tolist()
max_popped_iterations
```

```python
import random
chosen_data = refined_data[refined_data['iteration'] == random.choice(allocated_iterations)][['iteration','seq','angle','throwSec']]
throwData = chosen_data[["angle","throwSec"]].values.tolist()
throwData
```
